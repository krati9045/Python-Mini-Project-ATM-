1. INTRODUCTION 

   Automated Teller Machine enables the clients of a bank to have access to their account without going to the bank.
   Security is the foundation of the good ATM sysytem. This system will provide secure transactions between users and the bank      servers. The whole process will be automated right from PIN validation to transaction completion.
   
   To develop this ATM system the entire operation has been divided into the following step:
      --->verification process
      --->service selection
      --->Banking services
      --->Transactions

2. Different Modules/Functions Used

   __init__(self): It is called whenever a class object is created. self represents a class's instance and is required to                          access any variables or methods within the class.
   
   menu :  It is used to ask what the user want to do.It consists of various options like-->
         
         Check_Saving_Balance: It checks the current balance of the user on validation of the correct pin.
         
         Deposit_Funds       : It asks the user the amount he/she want to deposit into their account.
                               Updated balance = deposited balance + current saving balance 
                               It the updates the users current balance and shows the checking balance.
                               
         Withdraw_Funds      : It asks the user the amount he/she want to withdrwa from their account.
                               Updated balance = current saving balance - withdrawn balance
                               It the updates the users current balance and shows the checking balance.
                               
         Transfer_Funds      : It transfers amount from others account into self account.
                               It displays the balance of both the accounts, ie saving_balance and the 
                               checking balance.
3. Working :
            First the user is asked to enter his/her name and pin that are already stored in the dictionary.
            After successful validation of the pin the user is asked to select the available services by the ATM.
            To check the balance or deposit the balance or withdraw the balance or transfer funds, the user is again 
            asked to enter the pin.
            *********************This ensures 2-step verification for authenticated transaction*********************
            
            After this the selected choice of the user is performed and if something goes wrong the program terminates.
